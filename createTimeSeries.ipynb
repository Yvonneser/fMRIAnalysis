# Generate Timeserise npy Files
### Imports

```python
import nibabel as nib
import numpy as np
from matplotlib import pyplot as plt
```

# Read .mgz files and create a timeserise files

```python
filename = ('subj50_fs5_fwhm6_{}.mgh') # where {} will be replaced with 'lh' and 'rh'
timeseries = [None] * 2
for i, h in enumerate(['lh', 'rh']):
        timeseries[i] = nib.load(filename.format(h)).get_fdata().squeeze()
timeseries = np.vstack(timeseries)
np.save(('sub-50_TS.npy'), timeseries)
```

```python
plt.plot(timeseries[1000])
plt.show()
```
